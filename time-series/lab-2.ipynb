```python
data = pd.date_range(start='1/1/1956', end='1/1/1996', freq='M')
df['Time Stamp'] = pd.DataFrame(data)
df.to_csv('data/Gas-new.csv')
```

```python
df = pd.read_csv('data/Gas-new.csv', parse_dates=['Time Stamp'], index_col=['Time Stamp'])
df = df.drop('Unnamed: 0', axis=1)
print(df.info())
df.plot(figsize=(12, 6), grid=True)
plt.show()

df_sd = seasonal_decompose(df, model='additive', period=12)
df_sd.plot()
plt.show()
```
